```python
### Q15 type of computing platform used in data science projects?(VIA PIE CHART)
index=Q['Q11'].value_counts().index.tolist()
value=Q['Q11'].value_counts().values.tolist()
value_dict=dict(zip(index,value))
value_dict={key: value for key , value in value_dict.items() if (key != 'None') & (key != 'Other')}
value_dict
```

```python
###Q:15: via seaborn()common using platforms

fig,ax1=plt.subplots(figsize=(20,10))
ax1.pie(value_dict.values(),labels=value_dict.keys(),pctdistance=0.8,autopct='%1.1f%%',textprops={'fontsize':16},startangle=90)
centre_circle=plt.Circle((0,0),0.6)
ax1.set_title("computing platforms used in projects",fontsize=26)
plt.style.use('seaborn-deep')
plt.show()
```

```python
### Q:16 another visualization via matplot (majority for laptops)
fig,ax1=plt.subplots(figsize=(20,10))   
ax1.pie(value_dict.values(),labels=value_dict.keys(),pctdistance=0.8,autopct='%1.1f%%',textprops={'fontsize':16},startangle=90)
centre_circle=plt.Circle((0,0),0.6)
ax1.set_title("computing platforms used in projects",fontsize=26)
plt.show()
```
