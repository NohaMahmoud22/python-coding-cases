```python
import json
from io import StringIO
import pandas as pd
records = [json.loads(line) for line in open(r'C:\Users\Omar\Desktop\task 1 python\Task 2\Task 2\usa.gov_click_data_1.json')]
```

```python
dataframe = pd.json_normalize(records)
dataframe
```

```python
df = dataframe.dropna()
df = df.reset_index(drop=True)
```

```python

df['web_browser'] =pd.Series([x.split()[0] for x in df.a.dropna()])
```

```python
df2 = df['a'].str.extract(r'\(([^()]+)\)')
```

```python
##geting opwearting system
df['operating_sys']=pd.Series([x.split(';')[0] for x in df2[0].dropna()])
```

```python
df
```

```python
df['from_url']=df['r'].str.extract('//([\w\-\.]+)/').fillna('direct')
#df['to_url']=df['u'].str.extract('//([\w\-\.]+)/')
```

```python
df['to_url']=df['u'].str.extract('//([\w\-\.]+)/')
```

```python
df
```

```python
df
```

```python
df.rename(columns={'c': 'Country'}, inplace=True)
df.rename(columns={'cy': 'City'}, inplace=True)
df.rename(columns={'tz': 'time zone'}, inplace=True)
df.rename(columns={'r': 'FROM_URL'}, inplace=True)
df.rename(columns={'u': 'TO_URL'}, inplace=True)
df.rename(columns={'ll': 'Longitude and Latitude'}, inplace=True)
df.rename(columns={'hc': 'Time_in'}, inplace=True)
df.rename(columns={'t': 'Time_out'}, inplace=True)
```

```python
df['Longtuide']=df['Longitude and Latitude'].apply(str).str.strip('[]').str.split(',', expand=True)[0]
df['Latituide']=df['Longitude and Latitude'].apply(str).str.strip('[]').str.split(',', expand=True)[1]
## splitting attitude and lattitude
```

```python
df
```

```python
df= df.drop(columns =['nk','gr','Longitude and Latitude','h','l','al','hh','g','FROM_URL','TO_URL'])
```
